```python
import heapq
from collections import deque

# Sample orders with (order_id, weight, priority)
orders = [
    {"id": 1, "weight": 10, "priority": 1},
    {"id": 2, "weight": 15, "priority": 2},
    {"id": 3, "weight": 5, "priority": 1},
    {"id": 4, "weight": 8, "priority": 3},
    {"id": 5, "weight": 12, "priority": 2}
]

# Sample vehicles with (vehicle_id, capacity, available)
vehicles = [
    {"id": "V1", "capacity": 25, "available": True},
    {"id": "V2", "capacity": 20, "available": True}
]

# Step 1: Sort orders by priority (lower number = higher priority)
orders.sort(key=lambda x: x["priority"])

# Step 2: Assign orders to vehicles based on capacity
assigned_orders = {v["id"]: [] for v in vehicles}  # Dictionary to track vehicle assignments
unassigned_orders = deque()  # Queue for unassigned orders

for order in orders:
    assigned = False
    for vehicle in vehicles:
        total_weight = sum(o["weight"] for o in assigned_orders[vehicle["id"]])
        if vehicle["available"] and total_weight + order["weight"] <= vehicle["capacity"]:
            assigned_orders[vehicle["id"]].append(order)
            assigned = True
            break
    if not assigned:
        unassigned_orders.append(order)  # Queue unassigned orders

# Step 3: Dynamic Adjustments for unassigned orders
while unassigned_orders:
    order = unassigned_orders.popleft()
    for vehicle in vehicles:
        total_weight = sum(o["weight"] for o in assigned_orders[vehicle["id"]])
        if total_weight + order["weight"] <= vehicle["capacity"]:
            assigned_orders[vehicle["id"]].append(order)
            break
    else:
        print(f"Order {order['id']} cannot be assigned.")

# Output the assignment results
for vehicle, orders in assigned_orders.items():
    print(f"Vehicle {vehicle} assigned orders: {orders}")
```

```python
import heapq
from collections import deque

class Order:
    def __init__(self, order_id, weight, priority):
        self.id = order_id
        self.weight = weight
        self.priority = priority

    def __lt__(self, other):
        return self.priority < other.priority  # Allows priority queue sorting

    def __repr__(self):
        return f"Order(id={self.id}, weight={self.weight}, priority={self.priority})"

class Vehicle:
    def __init__(self, vehicle_id, capacity):
        self.id = vehicle_id
        self.capacity = capacity
        self.assigned_orders = []

    def available_capacity(self):
        return self.capacity - sum(order.weight for order in self.assigned_orders)

    def assign_order(self, order):
        if self.available_capacity() >= order.weight:
            self.assigned_orders.append(order)
            return True
        return False

    def __repr__(self):
        return f"Vehicle(id={self.id}, capacity={self.capacity}, assigned_orders={self.assigned_orders})"

class OrderManager:
    def __init__(self, orders, vehicles):
        self.order_queue = []
        for order in orders:
            heapq.heappush(self.order_queue, order)  # Min-heap for priority queue
        self.vehicles = {v.id: v for v in vehicles}
        self.unassigned_orders = deque()

    def assign_orders(self):
        while self.order_queue:
            order = heapq.heappop(self.order_queue)
            assigned = False
            for vehicle in self.vehicles.values():
                if vehicle.assign_order(order):
                    assigned = True
                    break
            if not assigned:
                self.unassigned_orders.append(order)

    def reassign_unassigned_orders(self):
        while self.unassigned_orders:
            order = self.unassigned_orders.popleft()
            for vehicle in self.vehicles.values():
                if vehicle.assign_order(order):
                    break
            else:
                print(f"⚠️ Order {order.id} could not be assigned due to capacity constraints.")

    def display_assignments(self):
        for vehicle in self.vehicles.values():
            print(f"🚚 Vehicle {vehicle.id} assigned orders: {vehicle.assigned_orders}")

# Sample orders
orders = [
    Order(1, 10, 1),
    Order(2, 15, 2),
    Order(3, 5, 1),
    Order(4, 8, 3),
    Order(5, 12, 2)
]

# Sample vehicles
vehicles = [
    Vehicle("V1", 25),
    Vehicle("V2", 20)
]

# Order management
manager = OrderManager(orders, vehicles)
manager.assign_orders()
manager.reassign_unassigned_orders()
manager.display_assignments()
```
